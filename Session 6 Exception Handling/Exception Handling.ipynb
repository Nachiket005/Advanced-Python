### Try Except Block



```python
def divide_handled():
    try:
        a = float(input('Please Enter Numerator : '))
        b = float(input('Please Enter Denominator : '))
        div = a/b
        return div
    except Exception as e:
        print(f'Exception occured : {e}')
```

```python
divide_handled()
```

```python
divide_handled()
print(4*9)
```

```python
divide_handled()
```

```python
def div_hand2():
    try:
        a = float(input('Please Enter Numerator : '))
        b = float(input('Please Enter Denominator : '))
        div = a/b
        return div
    except ZeroDivisionError:
        print('Cannot divide by 0, give any other number in denominator')
    except ValueError:
        print('Please do not enter alphabets instead of numbers')
```

```python
div_hand2()
```

```python
div_hand2()
```

### Handling input exceptions

```python
try:
    age = int(input('Please enter your age : '))
    if age<0 or age >100:
        raise ValueError("Age should be between o and 100")
    print (f'Age of {age} years is VALID age')
except:
    print(f'Invalid age {age} was entered')
```

```python
def sqr_root(n):
    try:
        from math import sqrt
        s = sqrt(n)
        print(f'Square root of given {n} is {s:.4f}')
    except Exception as e:
        print(f'Exception Occured :{e}')
    finally:
        print(f'This block is always excuted Finally')
```

```python
sqr_root(10)
```

```python
sqr_root(-11)
```

### Quadratic Equation Solution

ax^2 + bx + c = 0



```python
def quadratic_roots(a,b,c):
    try:
        from math import sqrt
        d = b**2 - 4*a*c
        x1 = (-b+sqrt(d))/(2*a)
        x2 = (-b-sqrt(d))/(2*a)
        print(f'Real Roots of Equation found : {x1:.4f} and {x2:.4f}')
        return x1, x2
    except Exception as e:
        print(f'Exception Occured : {e}')
    finally:
        print('Thank you for using this function')
```

x^2 - 5x + 6 = 0

a = 1 b = -5 c = 6 

```python
quadratic_roots(1, -5, 6)
```

x^2 + x + 1 = 0

a = 1, b = 1, c = 1

```python
quadratic_roots(1, 1, 1)
```
