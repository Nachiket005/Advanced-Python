### Find out name of player who has hit max home runs

```python
%%sql 
SELECT MAX(HmRun) FROM hits;
```

```python
%%sql
SELECT Name From hits
WHERE HmRun = (SELECT MAX(HmRun) From Hits);
```

### Find out league wise total number of players

```python
%%sql
SELECT League, count(League) as PlayerNum FROM hits
GROUP BY League
```

```python
df.groupby(by='League').count()['Name']
```

```python
%%sql
SELECT Division, SUM(Salary) as totalSalary FROM hits
GROUP BY Division;
```

```python
df.groupby(by='Division').sum()['Salary']
```

### Find out five most senior players

```python
%%sql
SELECT Name, Years FROM hits
ORDER BY Years DESC
LIMIT 5;
```
